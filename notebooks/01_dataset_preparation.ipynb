```python
from datasets import load_dataset
from PIL import Image
import os

# Create folders
os.makedirs("data/raw/images", exist_ok=True)

# Load COCO 2017 train split
dataset = load_dataset(
    "detection-datasets/coco",
    "2017",
    split="train",
    trust_remote_code=True
)

# COCO category ID  name mapping (partial for your classes)
COCO_ID_TO_NAME = {
    1: "person",
    2: "bicycle",
    3: "car",
    4: "motorcycle",
    5: "airplane",
    6: "bus",
    7: "train",
    8: "truck",
    16: "bird",
    17: "cat",
    18: "dog",
    19: "horse",
    21: "cow",
    22: "elephant",
    44: "bottle",
    46: "wine glass",
    47: "cup",
    51: "bowl",
    54: "sandwich",
    55: "orange",
    56: "broccoli",
    57: "carrot",
    58: "hot dog",
    59: "pizza",
    60: "donut",
    61: "cake",
    62: "chair",
    63: "couch",
    65: "bed",
    64: "potted plant",
    10: "traffic light",
    13: "bench",
    11: "fire hydrant",
    12: "stop sign"
}

TARGET_CLASSES = set(COCO_ID_TO_NAME.values())

saved = 0
MAX_IMAGES = 100

for item in dataset:
    image = item["image"]
    annotations = item["annotations"]

    found = False
    for ann in annotations:
        class_name = COCO_ID_TO_NAME.get(ann["category_id"])
        if class_name in TARGET_CLASSES:
            found = True
            break

    if found:
        image.save(f"data/raw/images/img_{saved}.jpg")
        saved += 1

    if saved >= MAX_IMAGES:
        break

print("Images downloaded:", saved)

```
